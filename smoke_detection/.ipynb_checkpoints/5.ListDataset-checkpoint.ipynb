```python
# _getitem_(self, index)
import numpy as np
from PIL import Image
import glob

imgs = sorted(glob.glob(r'/home/lsc/a409/users/lisuicheng/Machine_learning/PyTorch-YOLOv3/data/samples/*'))
                    
imgs
```

```python
img = Image.open(imgs[1])
img.show()
img = np.array(Image.open(imgs[1]))
img 
```

```python
h,w, _ = img.shape
h, w, img.shape
```

```python
dim_diff = np.abs(h - w)
dim_diff
```

```python
pad1, pad2 = dim_diff // 2, dim_diff - dim_diff // 2 # // 2 对半分
pad1, pad2
```

```python
pad = ((pad1, pad2), (0, 0), (0, 0)) if h <= w else ((0, 0), (pad1, pad2), (0, 0))
pad
```

```python
from skimage.transform import resize

img_no_padding = Image.fromarray(img.astype('uint8'))
# img_no_padding.show() # (512, 773) 
# Add padding
input_img = np.pad(img, pad, 'constant', constant_values=127.5) / 255.
padded_h, padded_w, _ = input_img.shape
img_padding = Image.fromarray(input_img.astype('uint8')).convert('RGB')
# img_padding.show() # (773, 773) , 除了255后(归一化),就是一片黑
# Resize and normalize
print(type(input_img))
input_img = resize(input_img, (416, 416, 3), mode='reflect')
img_resize = Image.fromarray(input_img.astype('uint8'))
# img_resize.show() # (416, 416)
print(type(input_img))
# input_img
```

```python
import numpy as np

label_path = '/home/lsc/a409/users/lisuicheng/Machine_learning/PyTorch-YOLOv3/data/coco/labels/train2014/COCO_train2014_000000005482.txt'
labels = np.loadtxt(label_path)
print(labels)
labels = labels.reshape(-1, 5)
labels, labels.shape
```

```python
h,w, _ = img.shape

print(h, w, padded_h, padded_w) # h, w: 原image的shape, padded_h, padded_w: padding后的h, w

# Extract coordinates for unpadded + unscaled image
x1 = w * (labels[:, 1] - labels[:, 3]/2) # labels[:, 1]是anchor的中心坐标x, labels[:, 3]时anchor的底边
y1 = h * (labels[:, 2] - labels[:, 4]/2)
x2 = w * (labels[:, 1] + labels[:, 3]/2)
y2 = h * (labels[:, 2] + labels[:, 4]/2)

print(x1, y1, x2, y2) # 没padding前image的labels
print(labels[:, 1], labels[:, 3])
```

```python
# Adjust for added padding
x1 += pad[1][0] # x方向加上(左边的)padding
y1 += pad[0][0] # y方向加上(下边的)padding
x2 += pad[1][0]
y2 += pad[0][0]

print(x1, y1, x2, y2) # 加上后坐标
print(pad, pad[1][0])
# Calculate ratios from coordinates
print(labels)
labels[:, 1] = ((x1 + x2) / 2) / padded_w # 归一化(除以padding后的底和高) 
labels[:, 2] = ((y1 + y2) / 2) / padded_h
labels[:, 3] *= w / padded_w
labels[:, 4] *= h / padded_h

print(labels) # 最终归一化的坐标
```

```python
# Fill matrix
max_objects = 50
filled_labels = np.zeros((max_objects, 5))
if labels is not None:
    filled_labels[range(len(labels))[:max_objects]] = labels[:max_objects]

# print(filled_labels[range(len(labels))[:max_objects]]) # 取labels长度的迭代器
print(len(labels), range(len(labels))[:max_objects])
print(filled_labels[range(len(labels))[:max_objects]])
filled_labels = torch.from_numpy(filled_labels)
print(filled_labels)
```
