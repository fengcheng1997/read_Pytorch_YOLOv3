```python
def write_labels(csv_name):
    csv_data = pd.read_csv(csv_name)
    
    # 截取视频名字
    (filepath,tempfilename) = os.path.split(csv_name)
    (shotname,extension) = os.path.splitext(tempfilename)
    mov_name = shotname[:8] #截取视频名字,eg. DSC_0146
    
    print(mov_name)
    
    for i in range(len(csv_data)):
        x1_ = csv_data.iloc[i].x1 / 1920 # frame_width = 1920, 归一化
        y1_ = csv_data.iloc[i].y1 / 1080
        x2_ = csv_data.iloc[i].x2 / 1920
        y2_ = csv_data.iloc[i].y2 / 1080
        
        x1 = (x1_ + x2_) / 2  
        x2 = x2_ - x1_
        y1 = (y1_ + y2_) / 2
        y2 = y2_ - y1_
        with open('/home/lsc/a409/users/lisuicheng/Machine_learning/PyTorch-YOLOv3/data/smoke_detection/labels/train/%s_%s.txt' 
                  % (mov_name,csv_data.iloc[i].frame_idx), 'a') as f:
            print(mov_name, csv_data.iloc[i].frame_idx)
            f.writelines('0 %.6f %.6f %.6f %.6f\n' % (x1, y1, x2, y2))
            

for csv_name in csv_list:
    write_labels(csv_name)
```
